```python
##Loading and Cleaning the Data##
```

```python
#Loading the data
import pandas as pd

events = pd.read_csv('events1.csv')
items = pd.read_csv('items.csv')
```

```python
# A. convert dates--time-series analyis
events['date'] = pd.to_datetime(events['date'])
```

```python
# B. filter for purchases
#the dataset includes 'add_to_cart' and 'view_item"
#for dashboard isolate 'purchase' events

purchases = events[events['type'] == 'purchase'].copy()
```

```python
# C. handle missing values
#check for nulls in critical columns 'item_id' 'price_in_usd'
purchases = purchases.dropna(subset=['item_id'])
```

```python
# D. Join tables
#joining events with item metadata to get product names and categories
df = pd.merge(purchases, items, left_on='item_id', right_on='id', how='left')
```

```python
# E. removing duplicates
df = df.drop_duplicates()
```

```python
##Analyzing the Data##
```

```python
#Top selling products
# Top 10 most purchased products
top_products = df.groupby('name')['item_id'].count().sort_values(ascending=False).head(10)

print(top_products)
```

```python
#Regions (Countries) that generate most revenue
revenue_by_country = df.groupby('country')['price_in_usd'].sum().sort_values(ascending=False).head(10)

print(revenue_by_country)
```

```python
#Month with the highest sales
df['month'] = df['date'].dt.strftime('%Y-%m')
monthly_trend = df.groupby('month')['price_in_usd'].sum()

print(monthly_trend)
```
